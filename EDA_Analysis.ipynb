# **Project Summary -**

This project performs an in-depth Exploratory Data Analysis (EDA) on the Global
Terrorism Dataset with the objective of identifying patterns, trends, and key
insights related to terrorist incidents worldwide. The dataset contains detailed
information such as the year of the incident, location (country and region),
type of attack, target category, and the number of casualties involved.

The initial phase of the analysis involved data cleaning and preparation.
Missing values were handled using statistical techniques like median for
numerical variables and mode for categorical variables to preserve the natural
distribution of the data. Important features were selected to simplify the dataset,
and a new metric called 'impact_score' was introduced by combining fatalities
and injuries to measure the severity of each incident. Duplicate entries were
also removed to ensure data quality and reliability.

Various visualizations were developed to better understand the dataset.
The time-series analysis indicated a steady rise in terrorist activities over
the years, particularly after the early 2000s, reflecting an increase in global
security challenges. This trend highlights the need for continuous monitoring
and stronger preventive strategies.

Further analysis at the country level showed that terrorist incidents are not
uniformly distributed but are concentrated in specific nations. This suggests
the existence of high-risk areas that require focused attention. Similarly,
regional analysis revealed that areas such as the Middle East & North Africa
and South Asia experience a significantly higher number of incidents compared
to other parts of the world.

When analyzing attack methods, it was observed that a limited number of attack
types, especially bombings and explosions, account for a large proportion of
incidents. This indicates a pattern in the choice of attack strategies, which
can be useful for designing targeted countermeasures.

The distribution of casualties demonstrated that while most incidents result
in relatively low impact, a small number of events lead to extremely high
casualties. This uneven distribution emphasizes the importance of being prepared
for rare but severe incidents that can have major consequences.

From a business and strategic perspective, these findings are highly useful.
Organizations can leverage these insights to identify safer regions for operations,
enhance security protocols, and develop effective risk mitigation strategies.
At the same time, the increasing frequency of attacks and concentration in certain
regions may negatively affect economic growth by discouraging investments,
reducing tourism, and creating uncertainty in the business environment.

Overall, this analysis provides a clear understanding of global terrorism trends.
It demonstrates how data-driven insights can support better decision-making,
improve safety measures, and help organizations operate more effectively in
high-risk and uncertain conditions.

### Variables Description

The dataset contains multiple variables that describe different aspects of terrorist incidents:

1. iyear:
   Represents the year in which the terrorist attack occurred.

2. country_txt:
   Name of the country where the incident took place.

3. region_txt:
   Indicates the geographical region of the attack (e.g., South Asia, Middle East).

4. attacktype1_txt:
   Describes the type of attack carried out, such as bombing, armed assault, or hijacking.

5. targtype1_txt:
   Specifies the type of target involved, such as civilians, military, government, etc.

6. nkill:
   Number of people killed in the attack.

7. nwound:
   Number of people wounded in the attack.

8. total_casualties / impact_score:
   A derived variable calculated as the sum of killed and wounded, representing
   the overall impact of the incident.

These variables provide both categorical and numerical information, enabling
detailed analysis of patterns, trends, and severity of terrorist activities.

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready
df['country_txt'].fillna("Unknown", inplace=True)
df['region_txt'].fillna("Unknown", inplace=True)
df['attacktype1_txt'].fillna("Unknown", inplace=True)
df['targtype1_txt'].fillna("Unknown", inplace=True)

df['nkill'].fillna(0, inplace=True)
df['nwound'].fillna(0, inplace=True)

df['total_casualties'] = df['nkill'] + df['nwound']

df = df[['iyear', 'country_txt', 'region_txt',
         'attacktype1_txt', 'targtype1_txt',
         'nkill', 'nwound', 'total_casualties']]

df.drop_duplicates(inplace=True)

df['iyear'] = df['iyear'].astype(int)

df.head()
````

### What all manipulations have you done and insights you found?

Data preprocessing steps performed:

1. Handled missing values by replacing categorical nulls with 'Unknown'
   and numerical nulls with 0.
2. Created a new feature 'total_casualties' for better analysis.
3. Selected only relevant columns to simplify the dataset.
4. Removed duplicate records to maintain data integrity.
5. Ensured correct data types for analysis.

The dataset is now clean and ready for exploratory data analysis.

## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

#### Chart - 1

````python
# Chart - 1 visualization code
yearly_data = df.groupby('iyear').size()

plt.figure(figsize=(10,6))
plt.plot(yearly_data.index, yearly_data.values, marker='o')

plt.title("Trend of Terrorist Attacks Over Years")
plt.xlabel("Year")
plt.ylabel("Number of Attacks")

plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

A line chart was chosen because the data represents a time-based trend (year-wise attacks).
Line charts are best suited for showing how values change over a continuous period.

It helps in clearly visualizing the increase or decrease in terrorist activities over the years,
making it easier to identify patterns, peaks, and trends in the dataset.

##### 2. What is/are the insight(s) found from the chart?

The chart shows that terrorist attacks have generally increased over the years,
with a noticeable rise after the year 2000.

There are certain peak periods where the number of attacks is significantly high,
indicating times of increased global conflict and instability.

Overall, the trend highlights that terrorism has become more frequent over time,
especially in recent decades.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

#### Chart - 2

````python
# Chart - 2 visualization code
country_data = (df.groupby('country_txt').size().sort_values(ascending=False).head(10))

plt.figure(figsize=(10,6))
plt.barh(country_data.index, country_data.values)

plt.title("Top 10 Countries by Number of Attacks")
plt.xlabel("Number of Attacks")
plt.ylabel("Country")

plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

A horizontal bar chart was chosen because the data involves comparing
multiple categories (countries) based on the number of attacks.

This type of chart makes it easy to compare values across categories and
is especially useful when category names are long, ensuring better readability.

It clearly highlights the countries with the highest number of attacks,
making the differences more understandable.

while many other countries experience relatively fewer incidents.

A small number of countries account for the majority of attacks, indicating
